#Coloque Tijolos na Parede

Não, não é "mais um tijolo na parede", é apenas um problema sobre somar números.

Suponha que você tem uma parede com o formato de um triângulo, como a mostrada abaixo. A parede tem 9 linhas, e a i-ésima linha tem exatamente i tijolos, considerando que a linha mais acima é a 1ª e que a mais abaixo é a 9ª. Alguns tijolos são rotulados com um número, enquanto os demais estão em branco. Os tijolos rotulados aparecem apenas em linhas ímpares, e ocupam posições ímpares dentro das suas linhas.



UOJ_1426.png

O problema que você deve resolver consiste em rotular os tijolos em branco com números, de tal forma que a seguinte regra seja satisfeita: O número de um tijolo é igual à soma dos números dos dois tijolos abaixo dele. Obviamente, esta regra não é aplicada à 9ª linha. Todos os números devem ser inteiros.

Nota: O exemplo de entrada contém dois casos de teste. O primeiro dele corresponde à parede mostrada acima.
#Entrada

A primeira linha da entrada contém um inteiro N, indicando o número de casos de teste. Esta linha é seguida pelos casos de teste. Cada caso é descrito por 5 linhas. Essas linhas correspondem às linhas ímpares da parede, de cima para baixo, como descrito acima. Cada linha contém os números nos tijolos já rotulados da linha correspondente na parede, da esquerda para a direita, separados por um espaço em branco.

Você pode assumir que todo caso de teste é correto, isto é, existe uma solução para o problema descrito.

#Saída

Para cada caso de teste, a saída deve ser formada por 9 linhas descrevendo os números de todos os tijolos da parede. Assim, a i-ésima linha deve conter os números correspondentes aos i tijolos da i-ésima linha da parede, da esquerda para a direita, separados por um espaço.

### Resolve o problema dos números faltantes nas linhas ímpares




```
#===================================================
# Completa os números faltantes nas linhas ímpares
#===================================================
def completa_lacunas_linha(linha_anterior, linha_atual):
    # cria uma lista identica a linha a ser preenchida
    linha_nova = linha_atual.copy()

    indice = 1
    for i in range(len(linha_anterior)):
        j = i + 1
        # Calcula o elemento faltante e insere na posição correta
        # Fórmula: subtrai o elemento i da linha_anterior da soma dos elementos
        # i e i+1 da linha_atual, o resultado dividido por 2 é o elemento faltante
        linha_nova.insert(indice, int(( linha_anterior[i] - ( linha_atual[i] + linha_atual[j] )) / 2) )
        indice += 2
    return(linha_nova)
```

### Resolve o problema das linhas pares


```
#===================================================
# Calcula os numeros das linhas pares
#===================================================
def gera_linhas_pares(linha_impar_posterior):
  linha_par = list()

  # O elemento i da linha par é a Soma do elemento i com o elemento i+1 da 
  # linha ímpar posterior
  for i in range(len(linha_impar_posterior) - 1):
    linha_par.append(linha_impar_posterior[i] + linha_impar_posterior[i + 1])
  return(linha_par)
```

```
#===================================================
# Completa os números faltantes nas linhas ímpares
#===================================================
def completa_lacunas_linha(linha_anterior, linha_atual):
    # cria uma lista identica a linha a ser preenchida
    linha_nova = linha_atual.copy()

    indice = 1
    for i in range(len(linha_anterior)):
        j = i + 1
        # Calcula o elemento faltante e insere na posição correta
        # Fórmula: subtrai o elemento i da linha_anterior da soma dos elementos
        # i e i+1 da linha_atual, o resultado dividido por 2 é o elemento faltante
        linha_nova.insert(indice, int(( linha_anterior[i] - ( linha_atual[i] + linha_atual[j] )) / 2) )
        indice += 2
    return(linha_nova)


#===================================================
# Calcula os numeros das linhas pares
#===================================================
def gera_linhas_pares(linha_impar_posterior):
  linha_par = list()

  # O elemento i da linha par é a Soma do elemento i com o elemento i+1 da 
  # linha ímpar posterior
  for i in range(len(linha_impar_posterior) - 1):
    linha_par.append(linha_impar_posterior[i] + linha_impar_posterior[i + 1])
  return(linha_par)


#======================================
# Lê as entradas do teclado e armazena
#======================================
# Número de casos de uso
repeticoes = int(input())

for i in range(repeticoes):
    
    # Ler todas as linhas de entrada
    linha_1 = [int(input())]
    linha_3 = list(map(int, input().split()))
    linha_5 = list(map(int, input().split()))
    linha_7 = list(map(int, input().split()))
    linha_9 = list(map(int, input().split()))

    # Completa os números faltantes nas linhas ímpares
    linha_9 = completa_lacunas_linha(linha_7, linha_9)
    linha_7 = completa_lacunas_linha(linha_5, linha_7)
    linha_5 = completa_lacunas_linha(linha_3, linha_5)    
    linha_3 = completa_lacunas_linha(linha_1, linha_3)    

    # Gera as linhas pares
    linha_2 = gera_linhas_pares(linha_3)
    linha_4 = gera_linhas_pares(linha_5)
    linha_6 = gera_linhas_pares(linha_7)
    linha_8 = gera_linhas_pares(linha_9)

    # Formata a impressão da saída conforme requisíto do teste    
    print(*linha_1)
    print(*linha_2)
    print(*linha_3)
    print(*linha_4)
    print(*linha_5)
    print(*linha_6)
    print(*linha_7)
    print(*linha_8)
    print(*linha_9)
```
